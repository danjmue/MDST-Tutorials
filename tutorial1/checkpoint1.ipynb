<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.plot.scatter(x='EXPENDITURE_PER_STUDENT_8', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('8th Grade Expenditure Per Student')
plt.ylabel('8th Grade Math Score')
```

**Visualization 2**

```python
df.plot.scatter(x='AVG_READING_8_SCORE', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('8th Grade Reading Score')
plt.ylabel('8th Grade Math Score')
```

There seems to be a strong positive correlation between 8th Grade Reading Score and 8th Grade Math Score. Once again, this makes sense as states with students who are testing well in reading are likely also testing well in math. However, correlation does not equal causation, and it is unlikely that a state's average reading score is responsible for its average math score.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['EXPENDITURE_PER_STUDENT_8', 'AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
model.score(X_test, y_test)
```

```python
np.mean((model.predict(X_test)-y_test)**2)**0.5
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'EXPENDITURE_PER_STUDENT_8'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('Math Score 8th Grade')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'EXPENDITURE_PER_STUDENT_8'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('Math Score 8th Grade')
plt.title("Model Behavior on Testing Set")
```

<h2> Summary </h2>

To investigate the predictability of 8th Grade Math Score, I decided to look at the Instruction Expenditure per Student, as well as the 8th Grade Reading Score. The resulting R^2 value of the model was 0.7530595700094598 and the MSE was 4.03021090462264, implying that the model fit the data well. Backing this up, the plots of both the training and testing sets seem to show the success of the model as well. Overall, it looks like Instruction Expenditure per Student and 8th Grade Reading Score could both be used to predict 8th Grade Math Score.
